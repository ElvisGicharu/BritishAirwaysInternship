```python
# imports
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import os

# regex
import re


import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
df=pd.read_csv('data/reviews.csv')
df
```

```python
# column for verification
df['Verified']=df.reviews.str.contains('Trip Verified')
```

```python
df
```

```python
#for lemmatization of words we will use nltk library
lemma = WordNetLemmatizer()
reviews_data = df.reviews.str.strip("✅ Trip Verified |")
reviews_data
```

```python
#list to collect cleaned data
nltk.download('wordnet')

rev =[]

for x in reviews_data:
    x=re.sub('[^a-zA-Z]',' ',x)
    x=x.lower()
    x=x.split()
    x=[lemma.lemmatize(word) for word in x if word not in set(stopwords.words("english"))]
    x=" ".join(x)
    rev.append(x)
```

```python
df['reviews1']=rev
```

```python
df.head()
```

```python
# convert date to datetime format
def clean_date(date_str):
    date_str = re.sub('[^\w\s]', '', date_str)
    date_str = date_str.replace('st', '').replace('nd', '').replace('rd', '').replace('th', '')  # Remove ordinal suffixes
    date_str = date_str.replace(' Augu ', ' August ')
    # date_str = date_str.replace(' July ', ' Jul ')
    # date_str = date_str.replace('June', ' Jun ')
    # date_str = date_str.replace('April', ' Apr ')
    # date_str = date_str.replace('March', ' Mar ')
    # date_str = date_str.replace(' February ', ' Feb ')
    # date_str = date_str.replace('January', ' Jan ')
    # date_str = date_str.replace('December', ' Dec ')
    # date_str = date_str.replace('November', ' Nov ')
    # date_str = date_str.replace('October', ' Oct ')
    # date_str = date_str.replace('September', ' Sep ')
    date_obj = datetime.strptime(date_str, '%d %B %Y')
    return date_obj

df['date'] = df['date'].apply(clean_date)
```
