```python
import pandas as pd
import datetime
```

```python
data=pd.read_csv('./data/reviews.csv')
data.head()
```

```python
# Drop unwanted columns
data=data.drop(columns='Unnamed: 0')
```

```python
# Working on columns one by one
#  creating new columns that could be used
data['Verified']=data['reviews'].str.contains('Trip Verified')
```

```python
data.head()
```

```python
#  remove unwanted texts using replace
data['reviews']=data['reviews'].str.replace('✅ Trip Verified |','')
```

```python
data['reviews']=data['reviews'].str.replace('Not Verified |','')
```

```python
data
```

```python
# Cleaning unwanted data in date column
data['date']=data['date'].str.replace('th','')  .replace('st','').replace('rd','')
```

```python
data['date'] =pd.to_datetime(data['date'],format='mixed')
```

```python
data
```

```python
data.to_csv('data/dataFromCleaner.csv')
```
